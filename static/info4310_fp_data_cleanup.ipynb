```python
import numpy as np
import pandas as pd
import statistics
```

```python
df = pd.read_csv('boba_nyc.csv')
df.head()
```

```python
df['categories'][2]
```

```python
# make a col containing a list of category aliases for a given row.
# made by looping through list of dicts(entry in category column),
# gets value of each "alias", appends that to a list.
```

```python
full_cat_list = []

for i in df.index:

  list_of_dicts = eval(df['categories'][i])
  row_cat_list = []

  for d in list_of_dicts:
    row_cat_list.append(d["alias"])

  full_cat_list.append(row_cat_list)


full_cat_list[2]
```

```python
df["catlist"] = full_cat_list
df['catlist'][2]
```
